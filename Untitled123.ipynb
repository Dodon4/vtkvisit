```python
import re
import math
class read:
    def __init__(self,filename):
        self.filename = filename 
    def read_cvs(self):
        pass
    def read_sel(self):
        file = open(self.filename)
        values = file.read().split("\n")
        data = []

        for key in values:
            value = re.findall(r"[-+]?\d*\.\d+|\d+", key)

            if value != []:
                data.append(value)
        return data
    def read_cart(self):
        file = open(self.filename)
        self.values = file.read().split("\n")
        self.data = []

        for key in self.values:
            self.value = re.findall(r"[-+]?\d*\.\d+|\d+", key)

            if self.value != []:
                self.data.append(self.value)
        return self.data
```

```python
class plot:
    def __init__(self,size,data,col):
        self.data = data[col-1::18]
        self.size=size
        #self.column=column
    def hex_cornerX(self,centerX, i):
        self.centerX=centerX
        self.i=i
        self.angle_deg = 60 * i   + 30
        self.angle_rad = math.pi / 180 * self.angle_deg
        return self.centerX + self.size * math.cos(self.angle_rad) 
    
    def hex_cornerY(self,centerY, i):
        self.centerY=centerY
        self.i=i
        self.angle_deg = 60 * i   + 30
        self.angle_rad = math.pi / 180 * self.angle_deg
        return self.centerY + self.size * math.sin(self.angle_rad)
    
    def column(self,matrix, i):
        return [float(row[i]) for row in matrix]
    
    def rings(self,matrix):
        self.matrix=matrix
        return (1 + (1 + 4 * ((self.matrix) - 1) / 3)**(1/2)) / 2
    
    def plt(self):
        self.points=self.column(self.data,2)
        self.num=len(self.points)-149
        self.r=self.rings(self.num)
        
        self.stepX=self.size/2*math.cos(math.pi/3)
        self.stepY=self.size/2*math.sin(math.pi/3)
        self.centerX=0
        self.centerY=0
        self.shiftY=0
        self.shiftX=0

        self.ring=0
        self.width = self.size * 2
        self.vert = self.width * 3/4
        self.height =  self.width**(3/2)
        self.k=0
        with open('2d.vtk', 'w', encoding='utf-8') as f:
            f.write("# vtk DataFile Version 3.0\nvtk output\nASCII\nDATASET POLYDATA\nPOINTS "+str(6*self.num)+" float\n")#str(len(points))
            for j in range(int(self.r)):
                if(j==0):
                    f.write(str(self.hex_cornerX(self.centerX,0))+" "+str(self.hex_cornerY(self.centerY,0))+\
                        " 0.0\n"+str(self.hex_cornerX(self.centerX,1))+" "+str(self.hex_cornerY(self.centerY,1))+\
                        " 0.0\n"+str(self.hex_cornerX(self.centerX,2))+" "+str(self.hex_cornerY(self.centerY,2))+\
                        " 0.0\n"+str(self.hex_cornerX(self.centerX,3))+" "+str(self.hex_cornerY(self.centerY,3))+\
                        " 0.0\n"+str(self.hex_cornerX(self.centerX,4))+" "+str(self.hex_cornerY(self.centerY,4))+\
                        " 0.0\n"+str(self.hex_cornerX(self.centerX,5))+" "+str(self.hex_cornerY(self.centerY,5))+\
                        " 0.0\n")#+str(points[index])+" 0.0" + "\n")
                for index in range(j*3):  
                        f.write(str(self.hex_cornerX(self.centerX,0))+" "+str(self.hex_cornerY(self.centerY,0))+\
                                " 0.0\n"+str(self.hex_cornerX(self.centerX,1))+" "+str(self.hex_cornerY(self.centerY,1))+\
                                " 0.0\n"+str(self.hex_cornerX(self.centerX,2))+" "+str(self.hex_cornerY(self.centerY,2))+\
                                " 0.0\n"+str(self.hex_cornerX(self.centerX,3))+" "+str(self.hex_cornerY(self.centerY,3))+\
                                " 0.0\n"+str(self.hex_cornerX(self.centerX,4))+" "+str(self.hex_cornerY(self.centerY,4))+\
                                " 0.0\n"+str(self.hex_cornerX(self.centerX,5))+" "+str(self.hex_cornerY(self.centerY,5))+" 0.0\n")#+str(points[index])+" 0.0" + "\n")
                        f.write(str(-self.hex_cornerX(self.centerX,0))+" "+str(-self.hex_cornerY(self.centerY,0))+\
                                " 0.0\n"+str(-self.hex_cornerX(self.centerX,1))+" "+str(-self.hex_cornerY(self.centerY,1))+\
                                " 0.0\n"+str(-self.hex_cornerX(self.centerX,2))+" "+str(-self.hex_cornerY(self.centerY,2))+\
                                " 0.0\n"+str(-self.hex_cornerX(self.centerX,3))+" "+str(-self.hex_cornerY(self.centerY,3))+\
                                " 0.0\n"+str(-self.hex_cornerX(self.centerX,4))+" "+str(-self.hex_cornerY(self.centerY,4))+\
                                " 0.0\n"+str(-self.hex_cornerX(self.centerX,5))+" "+str(-self.hex_cornerY(self.centerY,5))+" 0.0\n")#+str(points[index])+" 0.0" + "\n")
                        self.centerX-=self.width/2
                        self.centerY+=self.vert+self.size/4#height/5#**2/3#-size**(3/2)#-size**(3/2)**2

                        if(index >=j):
                            self.centerX-=self.width/2
                            self.centerY-=self.vert+self.size/4#height/5#**2/3#-size**(3/2)#-size**(3/2)**3/2
                        if(index >= 2 * j):
                            self.centerX+=self.width/2
                            self.centerY-=self.vert+self.size/4#height/5#**2/3#-size**(3/2)#-size**(3/2)**3/2
                self.shiftX += self.width
                self.centerX=self.shiftX
                self.centerY=self.shiftY
            f.write("POLYGONS "+str(self.num)+" "+str(self.num*7)+"\n")
            for index in range(self.num):
                f.write("6 "+str(0+index*6)+" "+str(1+index*6)+" "+str(2+index*6)+" "+str(3+index*6)+" "+str(4+index*6)+" "+str(5+index*6)+"\n")
            f.write("\nCELL_DATA "+ str(self.num)+"\n\nSCALARS p float\n"+"\nLOOKUP_TABLE default\n")
            for j in range(int(self.r)):
                if(j==0):
                    f.write(str(self.points[self.k])+"\n")
                    self.k+=1
                for index in range(j*3):
                    f.write(str(self.points[self.k])+"\n")
                    f.write(str(self.points[self.k+3*j])+"\n")
                    self.k+=1
                self.k+=3*j
```
